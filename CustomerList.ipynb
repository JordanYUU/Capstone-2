```python
df = pd.read_csv("Customer_List.csv", header=None, delimiter='|')
df.head(10)

# Initially used df[0].str.split("|", expand = True) for column separation.
```

```python
# def erase_symbols(name):
#     # Replace unwanted characters (e.g., '&' with 'and')
#     name = name.replace('&', '.').replace('!', '').replace('^', '')
#     return name

# df['cleaned_name'] = df['name'].apply(erase_symbols)

def clean_all_columns(df):
    for column in df.select_dtypes(include=['object']).columns:
        df[column] = df[column].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s\:\.\-]', '', str(x)))
    return df

df = clean_all_columns(df)
```

```python
pd.options.display.max_rows = None
df = df.drop(columns=['index'])
df = df.reset_index(drop=True)
df
```

```python
def clean_phone_number(phone):
    phone = re.sub(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', phone)
    return phone

df['phone_num'] = df['phone'].apply(clean_phone_number)
df
```
